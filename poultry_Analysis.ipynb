```python
# Hypothesis 2: Farmer Perspective on the Impact of Climate Change on Welfare and Behavior of Poultry

# Perform chi-square tests between "Climate_Changes_Observed" and welfare/behavior variables
welfare_behavior_vars = ['Observed _Changes _Increased _Temperature', 'Observed _Changes _Decreased _Rainfall',
                         'Observed _Changes _ Extreme _Weather _Events',
                         'Impact _on _Water _Availability', 'Impact _on _Bird _Health _and _Productivity']
p_values = []
for var in welfare_behavior_vars:
    contingency_table = pd.crosstab(df['Climate_Changes _Observed'], df[var])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    p_values.append(p)
    print(f"p-value for {var}: {p}")
contingency_table
```


##### The scatter plot summarizes the statistical significance of the relationship between climate changes and various welfare/behavior variables. The y-axis represents the negative logarithm (base 10) of the p-values, which indicate the level of significance.

##### In this analysis, the p-values for "Observed _Changes _Increased _Temperature", "Observed _Changes _Decreased _Rainfall", and "Observed _Changes _ Extreme _Weather _Events" are all 1.0, indicating no significant association with climate changes. Similarly, the p-values for "Impact _on _Water _Availability" and "Impact _on _Bird _Health _and _Productivity" are 0.7502 and 0.8934, respectively, suggesting non-significant relationships with climate changes.

##### Overall, these results suggest that the observed changes in temperature, rainfall, extreme weather events, water availability, and bird health/productivity do not show a significant association with climate changes. It implies that there may be other factors or variables influencing these welfare/behavior aspects in poultry farming, rather than solely being driven by climate changes.
